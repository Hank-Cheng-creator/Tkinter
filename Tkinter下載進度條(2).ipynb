```python
import tkinter as tk
import requests

window = tk.Tk()

#設置下載進度條

window.title('下載進度條示範 Part2') #設置視窗標題名稱
window.geometry('700x180')

tk.Label(window, text='下載進度:').place(x=40, y=20)
canvas = tk.Canvas(window, width=600, height=16, bg="white")
canvas.place(x=40, y=50)

#下載按鈕函數
def download():
    
    url_str = " https://statics.teams.cdn.office.net/production-osx/1.4.00.11161/Teams_osx.pkg"
    
    response = requests.get(url_str, stream=True)
    # When stream=True is set on the request
    # this avoids reading the content at once into memory for large responses.
    
    chunk_size = 4096 #設置每次下載的塊大小
    
    content_size = int(response.headers['Content-length'])
    #從返回的response的headers中獲取文件大小
    
    #填充進度條
    fill_line = canvas.create_rectangle(1.5, 1.5, 0, 23, width=0, fill="green")
    
    raise_data = 600 / (content_size/chunk_size)
    #進度條每次增加量
    
    global n
    
    #將下載檔案儲存至Jupyter Notebook預設工作路徑
    with open("Teams_osx.pkg", 'wb') as f: #僅有檔名產生，但尚未有內涵。
        
        n=0
        
        for data in response.iter_content(chunk_size=chunk_size): #每4096Kb寫入一次，逐步確保進度條
            f.write(data) #正式寫入檔案內涵
            n = n + raise_data #遞增n值
            canvas.coords(fill_line, (0, 0, n, 60)) #coords()填滿canvas，第一個參數為起點、第二個參數為終點
            window.update() #讓視窗更新
            
#清空進度條
def clean():
    #清空進度條
    fill_line = canvas.create_rectangle(1.5, 1.5, 0, 23, width=0, fill="white")
    
    for t in range(200):
        canvas.coords(fill_line, (0, 0, n, 60))
        window.update()
        
#下載按鈕

btn_download1 = tk.Button(window, text = '清空進度條', command=clean)
btn_download1.place(x=130, y=80)

btn_download2 = tk.Button(window, text = '開始下載', command=download)
btn_download2.place(x=250, y=80)

btn_exit = tk.Button(window, text = '離開', command=window.destroy)
btn_exit.place(x=360, y=80)

window.mainloop()
```
